```python
## Cleaning up the massive files and only taking the variables that interest u
later_tables_folder = local_data_folder + "acs-source-data/acs-1-year_tables_2005-2017/"
older_tables = os.listdir(later_tables_folder)
older_tables = [x for x in older_tables if "ACS" in x]

out_folder = local_data_folder + "acs_cleaned_pt1_out/"

for i, csv in enumerate(older_tables):
    year = int('20'+ csv[4:6])
    table_id = csv[11:-13] ## Different IDs per topic
    csv_full_name = later_tables_folder + csv
    this_year_table = codes_df[codes_df.year == year]
    this_year_table = this_year_table[this_year_table.table == table_id]
    var_codes = this_year_table.code.values
    new_varnames = this_year_table['should be'].values
    
    if len(var_codes)>0:
        df_i = pd.read_csv(csv_full_name)
        df_i = df_i.drop(0, axis = 0, inplace = False) ## First row not data
        dict_i = {'year': year,
                 'district': df_i['GEO.display-label'].values}
        for i, var_code in enumerate(var_codes):
            dict_i[new_varnames[i]] = df_i[var_code].values
        out_df = pd.DataFrame(dict_i)
        out_df.to_csv(out_folder+str(year)+"_"+table_id+".csv", index = False)
```

```python
## Combining ACS tables by year
cleaned_csv_folder = out_folder

this_folder_csvs = [x for x in os.listdir(cleaned_csv_folder) if "csv" in x]
df = pd.read_csv(cleaned_csv_folder + this_folder_csvs[0])

years = [int(x[:4]) for x in this_folder_csvs]
for i, csv in enumerate(this_folder_csvs[1:]):
    i = i+1 ## I'm starting with the 1st index of the list
    if years[i]>years[i-1]: ## If it's a new year
        df.to_csv(cleaned_csv_folder +"by year/"+ str(years[i-1]) + "_combined.csv",
                 index = False)
        df = pd.read_csv(cleaned_csv_folder+csv)
    else:
        df_i = pd.read_csv(cleaned_csv_folder + csv)
        df_i = df_i.drop('year', axis = 1, inplace =False)
        df = pd.merge(df, df_i, how = 'left',
                  on = "district")
    if i == len(this_folder_csvs)-1:
        df.to_csv(cleaned_csv_folder +"by year/"+ str(years[i-1]) + "_combined.csv",
                 index = False)
```

```python
## Concatenating the years together
year_csvs_folder = cleaned_csv_folder +"by year/"
year_csvs = [x for x in os.listdir(year_csvs_folder) if "csv" in x]

year_df = pd.read_csv(year_csvs_folder + year_csvs[0])
for csv in year_csvs[1:]:
    year_df_i = pd.read_csv(year_csvs_folder + csv)
    year_df = pd.concat([year_df, year_df_i], ignore_index=True)
    
df = year_df
```

```python
import re
```

```python
## Getting district numbers
district_nums = []

for i in range(len(df)):
    regex_match = re.search(" ([0-9]+) ", df.district[i])
    if regex_match is None:
        district_num_i = 1 ## At-large districts
    else:
        district_num_i = int(regex_match.group()[1:-1])
    district_nums.append(district_num_i)
df['district_num'] = district_nums
```

```python
## Getting state names
state_names = [x.partition(", ")[-1] for x in list(df.district.values)]
df['state_name'] = state_names

df = df.drop('district', axis =1, inplace = False)
```

```python
## Output cleaned csv
df.to_csv(local_data_folder + "ACS_2005_2017.csv", index = False)
```
